## Pruebas de hipótesis

**Planteamiento de hipótesis para la prueba t de Student de dos muestras independientes**

Hipótesis nula (H0):

La duración media de los viajes los sábados es la misma independientemente de si hay lluvia o buen clima.

Hipótesis alternativa (H1):

La duración media de los viajes los sábados es diferente si hay lluvia o buen clima.

### Identificación y modificación de los outliers

```python
# Valores promedio de los viajes con buen clima y mal clima
df7_raind_mean = df7.query("weather_conditions == 'Bad'")['duration_seconds'].mean().round(decimals = 3)
print(df7_raind_mean)
df7_goodd_mean = df7.query("weather_conditions != 'Bad'")['duration_seconds'].mean().round(decimals = 3)
print(df7_goodd_mean)
```

```python
# Determinación de outliers en la muestra para datos de viajes con mal clima
df7_raind = df7.query("weather_conditions == 'Bad'").reset_index()

columna_numerica = "duration_seconds" 

z_scores = np.abs(df7_raind[columna_numerica] - df7_raind[columna_numerica].mean()) / df7_raind[columna_numerica].std()

tolerancia = 3 

outliers = df7_raind[z_scores > tolerancia][columna_numerica]

print("Valores outliers en la columna", columna_numerica, ":")
print(outliers)
```

```python
# Se pone el valor de la media del conjunto de datos en el outlier
m_index = 41
df7_raind.loc[m_index, 'duration_seconds'] = df7_raind_mean
```

```python
# Determinación de outliers en la muestra para datos de viajes con buen clima
df7_goodd = df7.query("weather_conditions == 'Good'").reset_index()

columna_numerica = "duration_seconds" 

z_scores = np.abs(df7_goodd[columna_numerica] - df7_goodd[columna_numerica].mean()) / df7_goodd[columna_numerica].std()

tolerancia = 3

outliers = df7_goodd[z_scores > tolerancia][columna_numerica]

print("Valores outliers en la columna", columna_numerica, ":")
print(outliers)
```

```python
# Modificación de los valores de los outliers para poner valores medios del conjunto 
m_index = [70,302,489,518,540]
df7_goodd.loc[m_index, 'duration_seconds'] = df7_goodd_mean
```

```python
# Generación de conjuntos para día lluvioso y no lluvioso
df7_raind_values = df7_raind['duration_seconds']
df7_goodd_values = df7_goodd['duration_seconds']
alpha = 0.05

results = st.ttest_ind(df7_raind_values, df7_goodd_values, equal_var=True)

print('valor p:', results.pvalue)


if results.pvalue < alpha: 
    print("Rechazamos la hipótesis nula, existe una diferencia en la duración de los viajes los sábados dependiendo de si hay lluvia o buen clima")
else:
    print("No podemos rechazar la hipótesis nula, no existe diferencia entre la duración de los viajes los días sábado con o sin lluvia")

print('La duración media de los viajes el día sábado con lluvia es de {} '
      .format(df7_raind_values.mean().round(decimals=2)))
print('La duración media de los viajes el día sábado con buen clima  es de {} '
      .format(df7_goodd_values.mean().round(decimals=2)))
```

```python
df7_raind_values.plot(kind='hist', alpha =0.5, density= True, bins = 20, label= 'Lluvia', ax=plt.gca())
df7_goodd_values.plot(kind='hist', alpha= 0.5, density= True, bins =20, label='Buen clima', ax=plt.gca())
plt.title('Comparación de la distribución de la duración de los viajes los días sábado con lluvia o buen clima')
plt.legend()
plt.show()
```

### Conclusión de la sección
Existe una diferencia significativa en la duración de los viajes los sábados dependiendo de si hay lluvia o buen clima, ya que la duración media de los viajes los sábados con lluvia es considerablemente mayor que la duración media de los viajes los sábados con buen clima (2413.02 vs 1979.38). Los resultados de la prueba t de Student proporcionan evidencia contundente para rechazar la hipótesis nula y concluir que la duración de los viajes los sábados se ve afectada significativamente por las condiciones climáticas. Los viajes tienden a ser más largos los sábados con lluvia en comparación con los sábados con buen clima.
